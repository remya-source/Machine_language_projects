## Efficient Home Loan Approval: Enhancing Access and Speed

### <b><u> Problem Description:- </u></b>

The home loan approval process is often plagued by inefficiencies and delays, primarily due to manual assessment methods that rely on subjective evaluations of applicant creditworthiness. This traditional approach can lead to inconsistencies and prolonged decision times. To address these challenges, there is a growing need for a more streamlined and data-driven approach to home loan approvals. By leveraging advanced analytics and machine learning techniques, lenders can improve the accuracy and speed of credit assessments. Automated systems can analyze a wide range of financial data and applicant information to make more objective lending decisions, thereby reducing operational costs and enhancing customer satisfaction through faster response times.

### <b><u> Background:- </u></b>

Home loan approval hinges on accurately assessing an applicant's creditworthiness to minimize the risk of default. When a borrower defaults on a home loan, they fail to meet the scheduled repayment obligations, which can have profound financial repercussions for both the borrower and the lending institution. Predicting the likelihood of default is crucial for lenders to manage risk effectively and make informed lending decisions. By utilizing advanced analytics and predictive modeling, banks can analyze various factors such as credit history, income stability, debt-to-income ratio, and property value to assess risk more accurately. This proactive approach not only helps in mitigating potential losses but also ensures fair and consistent lending practices, ultimately benefiting both lenders and borrowers alike.

### Economic Impact of Default on Home Loan Applicants and Lenders

#### 1. Financial Stress on Borrowers:
  * **Increased debt burden:** Defaulting can lead to higher overall debt obligations for borrowers.
  * **Potential loss of home or property:** Non-payment risks foreclosure or loss of the property securing the loan.
  * **Impact on personal finances and future borrowing ability:** Defaults can impair credit scores, affecting future loan       
    eligibility.

#### 2. Losses for Lenders:
  * **Reduction in loan principal and interest income:** Defaults result in losses of expected income from both principal and   
      interest.
  * **Costs associated with foreclosure or repossession:** Lenders incur expenses related to legal proceedings and property 
      recovery.
  * **Operational expenses related to default management and recovery:** Additional costs arise from managing and recovering 
      defaulted loans.

#### 3. Wider Economic Effects:
  * **Impact on local housing market and property values:** Defaults can depress property values and affect local real estate  
      markets.
  * **Ripple effects on related industries(e.g., construction, real estate):** Default impacts can extend to industries closely 
      tied to housing.
  * **Overall economic stability and consumer confidence:** High default rates can undermine economic stability and impact 
      consumer spending and confidence.

## 1. Data Description

The dataset used for this research consists of a binary response variable, `Loan_Status`, with values (Yes = 1, No = 0). The study utilizes the following 23 explanatory variables:

12. **Property_Area:** Urban/ Semi Urban/ Rural.

13. **Loan_Status:** Loan approved (0 = No, 1 = Yes).

This dataset offers essential insights for analyzing home loan approval patterns and constructing predictive models to evaluate creditworthiness.

# 3. Feature Engineering

### Creating  a pipeline for preprocessing steps like encoding and scaling

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
import pandas as pd

def create_custom_pipeline(categorical_cols, numerical_cols, scaling_method=None):
    categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(drop="first", handle_unknown='ignore'))])

    if scaling_method:
        if scaling_method == 'robust':
            numerical_transformer = Pipeline(steps=[('scaler', RobustScaler())])
        else:
            numerical_transformer = "passthrough"
    else:
        numerical_transformer = "passthrough"

    preprocessor = ColumnTransformer(transformers=[('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)])

    pipeline = Pipeline(steps=[('preprocessor', preprocessor)])
    return pipeline
```

```python
print("Columns in home:", x_train.columns)

categorical_cols = ['Gender','Education', 'Married', 'Self_Employed', 'Property_Area', 'Dependents']
numerical_cols   = [col for col in x_train.columns if col not in categorical_cols]

print("Categorical Columns:", categorical_cols)
print("Numerical Columns:", numerical_cols)

pipeline = create_custom_pipeline(categorical_cols, numerical_cols, scaling_method='robust')

X_preprocessed_train = pipeline.fit_transform(x_train)
X_preprocessed_test  = pipeline.transform(x_test)
```
